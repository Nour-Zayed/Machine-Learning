```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
def gradient_descet_fr(x,y,iterations):
  ## assume random val for m , b
  m = 0
  b = 0
  lr = .01
  n = len(x)

  for i in range(iterations):
    y_hat =  m * x + b

    dedm = -(2/n) * sum(x * (y - y_hat))
    dedb = -(2/n) * sum(y - y_hat)

    m = m - lr * dedm
    b = b - lr * dedb

    print(sum(y-y_hat))
```
