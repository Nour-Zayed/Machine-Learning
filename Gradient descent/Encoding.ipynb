```python
import pandas as pd
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
df.replace(to_replace=['monroe township','west windsor','robinsville'], value=["Faysal", "El Helmya","Haram"], inplace=True)
```

```python
df
```

```python
df.groupby("town").mean()
```

```python
df.groupby("town").price.mean()
```

```python
df["price/sq"] = df.price/ df.area
df
```

```python
df.groupby("town").mean()[["price/sq"]]
```

```python
encodeed_towns =pd.get_dummies(df.town).astype(int)
encodeed_towns
```

```python
final_data = pd.concat([df,encodeed_towns], axis = 1)
final_data
```

```python
final_data.drop("town",axis= 1 ,inplace=True)
final_data
```

```python
final_data.drop("price/sq",axis= 1 ,inplace=True)
final_data
```

```python
x = final_data.drop('price', axis = 1)
y = final_data.price
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x,y)
```
