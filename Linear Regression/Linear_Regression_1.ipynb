```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
df.area.min()
```

```python
df.area.max()
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price)
plt.show()
```

```python
area = df.drop('price',axis='columns')
print(type(area))
area
```

```python
price = df.price
print(type(price))
price
```

```python
lin_regression = linear_model.LinearRegression()
lin_regression.fit(area,price)
```

#### (1) Predict price of a home with area = 3300 sqr ft

```python
lin_regression.predict([[3300]])
```

```python
lin_regression.predict([[3300]])[0]
```

```python
round((lin_regression.predict([[3300]])[0]),2)
```

```python
lin_regression.predict([[3200], [3300] , [4200]])
```

```python
lin_regression.predict([[2600]])
```

```python
# calculate accuracy
lin_regression.score(area,price)
```

```python
# y = mx + b
# when training , we try to find best values for m and b

```

```python
lin_regression.coef_ # m
```

```python
lin_regression.intercept_ #b
```
